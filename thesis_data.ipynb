Combine all gene data together.

```python
import pandas as pd
cambridge_all_list = list()
value = 0
for gsm_name, gsm in cambridge.gsms.items():
    #print("Name: ", gsm_name)
    print(gsm.metadata['characteristics_ch1'][0])
    if gsm.metadata['characteristics_ch1'][0][-1] == 'l':
        print('healthy')
        value = 0
    else:
        print('MDD')
        value = 1
    #print(gsm.table.tail())
    cambridge_all_list.append(
        gsm.table.append({'ID_REF':'TYPE','VALUE':value}, 
                         ignore_index=True))
```

```python
cambridge_all_list[0].tail()
```

```python
libr_all_list = list()
value = 0
for gsm_name, gsm in libr.gsms.items():
    #print("Name: ", gsm_name)
    print(gsm.metadata['characteristics_ch1'][0])
    if gsm.metadata['characteristics_ch1'][0][-1] == 'l':
        print('healthy')
        value = 0
    else:
        print('MDD')
        value = 1
    libr_all_list.append(
        gsm.table.append({'ID_REF':'TYPE','VALUE':value}, 
                         ignore_index=True).iloc[:,:2])
```

```python
libr_all_list[0].tail()
```

```python
japan_all_list = list()
value = 0
for gsm_name, gsm in japan.gsms.items():
    #print("Name: ", gsm_name)
    print(gsm.metadata['characteristics_ch1'][1])
    if gsm.metadata['characteristics_ch1'][1][-1] == 'y':
        print('healthy')
        value = 0
    else:
        print('MDD')
        value = 1
    japan_all_list.append(
        gsm.table.append({'ID_REF':'TYPE','VALUE':value}, 
                         ignore_index=True).iloc[:,:2])
```

```python
cambridge_all_list_reset_index = []
for i in cambridge_all_list:
    cambridge_all_list_reset_index.append(i.set_index('ID_REF'))
camp_all = pd.concat(cambridge_all_list_reset_index, axis=1)
camp_all.columns = [i for i in range(camp_all.shape[1])]
camp_all = camp_all.T
camp_all.to_pickle('camp_all')
```

```python
libr_all_list_reset_index = []
for i in libr_all_list:
    libr_all_list_reset_index.append(i.set_index('ID_REF'))
libr_all = pd.concat(libr_all_list_reset_index, axis=1)
libr_all.columns = [i for i in range(libr_all.shape[1])]
libr_all = libr_all.T
libr_all.to_pickle('libr_all')
```

```python
japan_all_list_reset_index = []
for i in japan_all_list:
    japan_all_list_reset_index.append(i.set_index('ID_REF'))
japan_all = pd.concat(japan_all_list_reset_index, axis=1)
japan_all.columns = [i for i in range(japan_all.shape[1])]
japan_all = japan_all.T
japan_all.to_pickle('japan_all')
```

```python
pd.concat([camp_all, libr_all, japan_all]).dropna()
```

```python
X = camp_all.iloc[:,:-1]
y = camp_all.iloc[:,-1]
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold=(.8 * (1 - .8)))
X_filter = sel.fit_transform(X)
X_filter_df = pd.DataFrame(X_filter)
X_filter_df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
                                    X_filter_df, y, test_size=0.33, random_state=42, stratify=y)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.decomposition import PCA

estimators = [('reduce_dim', PCA()), 
              ('clf', SVC())]
pipe = Pipeline(estimators)
pipe.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
print('accuracy_score:',accuracy_score(y_test, pipe.predict(X_test)))
print('confusion_matrix:',confusion_matrix(y_test, pipe.predict(X_test)))
print('f1_score:',f1_score(y_test, pipe.predict(X_test)))
```

use kfold to analyze model

```python
X = camp_all.iloc[:,:-1]
y = camp_all.iloc[:,-1]
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold=(.8 * (1 - .8)))
X_filter = sel.fit_transform(X)
X_filter_df = pd.DataFrame(X_filter)
X_filter_df.shape

from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.decomposition import PCA, FastICA
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score

rskf = RepeatedStratifiedKFold()
estimators = [('reduce_dim', PCA(n_components=2, svd_solver='full')), 
              ('clf', SVC())]
pipe = Pipeline(estimators)
f1_score_list = []

for train_index, test_index in rskf.split(X_filter_df, y):
    X_train, X_test = X_filter_df.iloc[train_index], X_filter_df.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    pipe.fit(X_train, y_train)
    #print('accuracy_score:',accuracy_score(y_test, pipe.predict(X_test)))
    #print('confusion_matrix:',confusion_matrix(y_test, pipe.predict(X_test)))
    print('f1_score:',f1_score(y_test, pipe.predict(X_test)))
    f1_score_list.append(f1_score(y_test, pipe.predict(X_test)))

```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(f1_score_list)
plt.ylabel('f1_score_list')
plt.show()
```
