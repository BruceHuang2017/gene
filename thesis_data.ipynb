The data is from GEO(Gene Expression Omnibus). 

1.Cambridge(GSE98793)

2.LIBR(GSE39653)

3.Japan(GSE76826)

In the dataset, sample is assigned to unique and stable GEO accession number that is composed of 'GSM'

```python
import GEOparse
cambridge = GEOparse.get_GEO(geo="GSE98793", destdir="./")
libr = GEOparse.get_GEO(geo="GSE39653", destdir="./")
japan = GEOparse.get_GEO(geo="GSE76826", destdir="./")
```

```python
# present a sample of what the data looks like. 
print("GSM example:")
for gsm_name, gsm in japan.gsms.items():
    print("Name: ", gsm_name)
    print("Metadata:",)
    for key, value in gsm.metadata.items():
        print(" - %s : %s" % (key, ", ".join(value)))
    print ("Table data:",)
    print (gsm.table.head())
    break
```

```python
for gsm_name, gsm in cambridge.gsms.items():
    print("Name: ", gsm_name)
    print(gsm.metadata['characteristics_ch1'])
```

Combine all gene data together.

```python
import pandas as pd
cambridge_all_list = list()
value = 0
for gsm_name, gsm in cambridge.gsms.items():
    #print("Name: ", gsm_name)
    print(gsm.metadata['characteristics_ch1'][0])
    if gsm.metadata['characteristics_ch1'][0][-1] == 'l':
        print('healthy')
        value = 0
    else:
        print('MDD')
        value = 1
    #print(gsm.table.tail())
    cambridge_all_list.append(
        gsm.table.append({'ID_REF':'TYPE','VALUE':value}, 
                         ignore_index=True))
```

```python
cambridge_all_list[0].tail()
```

```python
libr_all_list = list()
value = 0
for gsm_name, gsm in libr.gsms.items():
    #print("Name: ", gsm_name)
    print(gsm.metadata['characteristics_ch1'][0])
    if gsm.metadata['characteristics_ch1'][0][-1] == 'l':
        print('healthy')
        value = 0
    else:
        print('MDD')
        value = 1
    libr_all_list.append(
        gsm.table.append({'ID_REF':'TYPE','VALUE':value}, 
                         ignore_index=True).iloc[:,:2])
```

```python
libr_all_list[0].tail()
```

```python
japan_all_list = list()
value = 0
for gsm_name, gsm in japan.gsms.items():
    #print("Name: ", gsm_name)
    print(gsm.metadata['characteristics_ch1'][1])
    if gsm.metadata['characteristics_ch1'][1][-1] == 'y':
        print('healthy')
        value = 0
    else:
        print('MDD')
        value = 1
    japan_all_list.append(
        gsm.table.append({'ID_REF':'TYPE','VALUE':value}, 
                         ignore_index=True).iloc[:,:2])
```

```python
japan_all_list[0].tail()
```

```python
import pickle
with open('cambridge_all_list.pkl', 'wb') as f:
    pickle.dump(cambridge_all_list, f)
    
```
